# visualization

```python
mail['the'].plot()

plt.show
```

```python
plt.title("frequency of THE ")
plt.ylabel('the')
plt.xlabel('Email No.')
plt. xticks(rotation=45)
x = mail['Email No.']
y = mail['the']
plt.plot(x,y)
plt.show()
```

```python
plt.title('frequency of THE ')
plt.ylabel('the')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(mail)):
    x = mail['Email No.'][i]
    y = mail['the'][i]
    plt.plot(x, y, 'bo')
    if mail['the'][i] > 100:
        plt.text(x, y * (1 - 0.05), mail['Email No.'][i], fontsize=12)
    
plt.show()
```
